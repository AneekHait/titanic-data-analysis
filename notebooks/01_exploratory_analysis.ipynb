## Key Insights

1. **Only ~38% of passengers survived** (1,309 total).  
2. **Gender mattered most** — women had a 72.8% survival rate vs 19.1% for men.  
3. **Class hierarchy** — 1st class 62.0%, 2nd 42.8%, 3rd 25.5%.  
4. **Children (0-16)** had 49.0% survival rate.  
5. **Higher fare = higher survival** — top quintile ($42-$512) survived at 62.2%.  
6. **Embarkation port C** (Cherbourg) passengers fared best at 56.6%.  
7. **Age has only 51 missing values (3.9%)** — far better than Kaggle's 19.9%.  
8. **Fare-survived correlation (0.247)** is the strongest numerical predictor.
