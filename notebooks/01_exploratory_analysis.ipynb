## Key Insights

1. **Only ~38% of passengers survived.**  
2. **Gender mattered most** — women had a ~74% survival rate vs ~19% for men.  
3. **Class hierarchy** — 1st class ~63%, 2nd ~47%, 3rd ~24%.  
4. **Children (age < 10)** had higher survival odds.  
5. **Higher fare = higher survival** — those paying more were in better cabins/lifeboat access.  
6. **Embarkation port C** (Cherbourg) passengers fared better, likely correlated with class.  
7. **Age (177 missing) and Cabin (687 missing)** need imputation or dropping for modeling.
